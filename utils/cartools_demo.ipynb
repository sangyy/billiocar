```python
from lib.cartools import *

import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_adjust_axes(adjust_point=(1,1), color="#0033CC", reverse=False):
    xVal = np.linspace(-1,1,1001)
    yVal = []
    for i in xVal: 
        yVal.append(adjust_axes(val=i, adjust_point=adjust_point)) #这里改公式
    if reverse==True:
        yVal.reverse()

    plt.plot(xVal,yVal,'b-,',linewidth=1, color=color)

    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data',0))
```

```python
draw_adjust_axes(adjust_point=(1,0.5), color="#0033CC", reverse=False)
# draw_adjust_axes(adjust_point=(1,1), color="#FF0000", reverse=True)
```

```python
draw_adjust_axes(adjust_point=(0.8,0.4), color="#0033CC", reverse=False)
draw_adjust_axes(adjust_point=(0.8,0.4), color="#FF0000", reverse=True)
```

```python
def draw_comb(adjust_point=(1,1), oy=1, ny=-1, color="#0033CC", reverse=False):
    xVal = np.linspace(-1,1,1001)
    xAdj = []
    yVal = []
    for i in xVal: 
        xAdj.append(adjust_axes(val=i, adjust_point=adjust_point))
    for i in xAdj:
        yVal.append(steer_val(val=i, oy=oy, ny=ny))
    if reverse==True:
        yVal.reverse()

    plt.plot(xVal,yVal,'b-,',linewidth=1, color=color)

    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data',0))
```

```python
draw_comb(adjust_point=(0.9,0.5), oy=1, ny=-0.6, color="#0033CC", reverse=False)
draw_comb(adjust_point=(0.9,0.5), oy=1, ny=-0.6, color="#FF0000", reverse=True)
```
